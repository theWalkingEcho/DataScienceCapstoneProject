```python
# Task 2: Mark the success/failed launches for each site on the map
launch_data = pd.DataFrame({
    'Launch Site': ['CCAFS LC-40', 'CCAFS SLC-40', 'KSC LC-39A', 'VAFB SLC-4E', 'CCAFS LC-40'],
    'Lat': [28.562302, 28.563197, 28.573255, 34.632834, 28.562302],
    'Long': [-80.577356, -80.576820, -80.646895, -120.610745, -80.577356],
    'Class': [1, 0, 1, 0, 1]
})

# Center map on average coordinates
map_center = [launch_data['Lat'].mean(), launch_data['Long'].mean()]
site_map = folium.Map(location=map_center, zoom_start=5)

# Add markers for each launch
for index, row in launch_data.iterrows():
    coordinate = [row['Lat'], row['Long']]
    outcome = 'Success' if row['Class'] == 1 else 'Failure'
    color = 'green' if row['Class'] == 1 else 'red'

    folium.CircleMarker(
        location=coordinate,
        radius=6,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.6,
        popup=f"{row['Launch Site']} - {outcome}"
    ).add_to(site_map)

# Display the map
site_map
```

```python
import numpy as np
# TASK 3: Calculate the distances between a launch site to its proximities

ref_lat, ref_lon = 28.5721, -80.6480


# Haversine function to calculate distance (in kilometers)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    d_phi = np.radians(lat2 - lat1)
    d_lambda = np.radians(lon2 - lon1)

    a = np.sin(d_phi / 2) ** 2 + np.cos(phi1) * np.cos(phi2) * np.sin(d_lambda / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c


# Add a distance column to spacex_df
spacex_df['Distance_to_KSC'] = spacex_df.apply(
    lambda row: haversine(row['Lat'], row['Long'], ref_lat, ref_lon), axis=1
)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Haversine function to calculate distance (in kilometers)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    d_phi = np.radians(lat2 - lat1)
    d_lambda = np.radians(lon2 - lon1)

    a = np.sin(d_phi / 2) ** 2 + np.cos(phi1) * np.cos(phi2) * np.sin(d_lambda / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c

# Example: closest city coordinates (let's assume it's "City A")
city_lat, city_lon = 28.5600, -80.5900  # Replace with actual city coordinates

# Get coordinates of the launch site (Example: First launch site in spacex_df)
launch_site_lat = spacex_df.iloc[0]['Lat']
launch_site_lon = spacex_df.iloc[0]['Long']

# Calculate the distance between the launch site and the city
distance_city = haversine(launch_site_lat, launch_site_lon, city_lat, city_lon)

# Initialize the map centered on the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Create a marker for the city (closest point)
city_marker = folium.Marker(
    location=[city_lat, city_lon],
    popup=f"City A - Distance: {distance_city:.2f} km",
    icon=DivIcon(icon_size=(20, 20), icon_anchor=(0, 0), html=f'<div style="font-size: 12; color:#e74c3c;"><b>{distance_city:.2f} km</b></div>')
)

# Add the city marker to the map
city_marker.add_to(site_map)

# Draw a line from the launch site to the city
coordinates = [[launch_site_lat, launch_site_lon], [city_lat, city_lon]]
lines = folium.PolyLine(locations=coordinates, weight=2, color="green")
site_map.add_child(lines)

# Display the map
site_map
```
